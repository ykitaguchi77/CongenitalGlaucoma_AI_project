#**眼位写真NEW（年齢情報が付加されたファイルが入ったフォルダ）から、任意の年齢層を抜き出す**

```python
import os
import glob
import numpy as np
import shutil

path = glob.glob(r"H:\眼位写真NEW/*")
```

```python
#画像番号-日付-年齢-......jpgとなっている
os.path.basename(path[0])
```

```python
#年齢を抜き出し
os.path.basename(path[0]).split("-")[2]
```

```python
#10歳以下のリスト
u10_path_list = [x for x in path if int(os.path.basename(x).split("-")[2])<10]
u10_path_list
```

```python
u10_num_list = [int(os.path.basename(x).split("-")[0]) for x in u10_path_list]
u10_num_list = list(set(u10_num_list)) #重複を省略
u10_num_list.sort()

u10_list_numpy = np.array(u10_num_list) #リストを省略なしで表示させるためにnumpyの形式に直す
np.set_printoptions(threshold=np.inf) #numpyのリストを省略なしで表示
u10_list_numpy
```

#**該当するデータを新しいフォルダにコピー**

```python
dst_folder = f"H:/先天性緑内障/Under10_dataset"

os.makedirs(dst_folder, exist_ok = True)
```

```python
import time

# 時間計測開始
time_sta = time.perf_counter()

u10_path_list = [x for x in path if int(os.path.basename(x).split("-")[2])<10]

for orig_path in u10_path_list:
    dst_path = os.path.join(dst_folder, os.path.basename(orig_path))
    shutil.copy(orig_path, dst_path)

# 時間計測終了
time_end = time.perf_counter()
tim = time_end- time_sta
print(f"elapsed time: {tim}")
```
